## Coordinate Systems

All definitions and results are available in [Section 8.1](#ch8.1).

1. A **coordinate system** is a rule for uniquely assigning positions in a $d$-dimensional space to $d$ values

    - A **Cartesian coordinate system** specifies locations with a list of $d$ numbers, which each measure the distance from the location, $x$, to the origin, along a series of perpendicular directions. 

    - A **polar coordinate system** specifies locations in the plane with a distance from the origin, and an angle away from a reference direction (usually, the ray leaving the origin in the positive horizontal direction).

    - To convert between polar and Cartesian, use the rules:

    $$\text{Pol. to Cart. }\begin{cases} & x_1 = r \cos(\theta) \\ & x_1 = r \sin(\theta) \end{cases}, \quad \quad \text{Cart. to Pol. } \begin{cases} & r = \sqrt{x_1^2 + x_2^2} \\ & \theta = \tan^{-1}(x_2/x_1) \end{cases} $$



1. A **vector** is an ordered list of finitely many numbers, $v = [v_1,v_2,...,v_d]$. The number of entries is the dimension of the vector. 

    - Vectors represent positions in a Cartesian coordinate system. The **zero** vector is the vector $[0,0,...0]$. It points to the origin.

    - To **add** vectors, produce a new vector by adding their entries, $v + w = [v_1 + w_1, v_2 + w_2, ..., v_d + w_d]$. 

    - To **scale** a vector, multiply it by a **scalar**, $\lambda v = [\lambda v_1, \lambda v_2, ..., \lambda v_d]$. Scaling a vector preserves its direction, but changes its length. 

    - A **linear combination** of two vectors is a sum of the form $\alpha v + \beta w$ where $\alpha$ and $\beta$ are scalars, while $v$ and $w$ are vectors. 

    - The **length** or **magnitude** of a vector is defined $\|v\| = \sqrt{v_1^2 + v_2^2 + ... v_d^2}$. A **unit** vector is a vector of length 1.

    - The **direction** of a vector may be uniquely represented with the matching unit vector $\hat{v} = v/\|v\|$. Two vectors, $v$ and $w$, are **parallel** if and only if $$\hat{v} = \pm \hat{w}$. 

    - The angle, $\theta$, between two vectors may be measured using an **inner (dot) product**:

        $$v \cdot w = v_1 w_1 + v_2 w_2 + ... v_d w_d = \sum_{j=1}^d v_j w_j. $$

        - Geometrically:

        $$v \cdot w = \|v\| \|w\| \cos(\theta).$$

    - So, two vectors are **perpendicualr (orthogonal)** if and only if $v \cdot w = 0$. 




## Functions of Multiple Variables

All definitions and results are available in [Section 8.2](#ch8.2).

1. A **function of multiple variables** is a function that accpets multiple inputs (e.g. vectors). 

    - A function is **scalar-valued** if it returns scalars. It is **vector-valued** if it returns vectors. 

    - Scalar-valued functions of multiple variables define **surfaces**. 

1. We can visualize surfaces using:

    - **cross-sections**: which are slices of the surface produced by fixing all but one input, then varying the free input,

    - **level sets**: which are all inputs that produce a fixed output value, or

    - **heat maps**: which encode the output value as a color. 


## Joint Distributions

All definitions and results are available in [Section 8.3](#ch8.3).

1. Discrete random vectors are characterized by **joint probability mass functions** which may be represented using a **joint distribution table**

    - Marginal mass functions are produced by either summing across the rows, or down the columns

    - Conditional mass functions are produced by isolating a row or column, and scaling by the corresponding marginal to normalize. 

    - So, conditional mass functions are proportional to cross-sections of the table.

1. Continuous random vectors are characterized by **joint density functions** which are **surfaces**:

    - The joint density at $x = [x_1,x_2,...x_d]$ is the chance that every coordinate $X_j$ is within $\pm \frac{1}{2} \Delta x$ of $x_j$, scaled by the volume of the $d$-dimensional cube with side lengths $\Delta x$, in the limit as $\Delta x$ approaches zero. 

    - The rules for computing chances, finding marginal densities, finding conditional densities, checking independence, and evaluating expectations are *analogous* to all of the same procedures we developed in [Sections 1.5](#ch1.5) and [1.6](#ch1.6), as long as we work with densities instead of mass functions, and, replace all sums over possible values with integrals over possible vectors.
