# Coordinate Systems

To extend our modeling framework to multiple dimensions we will need to introduce language for describing mutlivariate problems. The most basic component of that language is a *coordinate system*.

:::{note} Coordinate Systems and Dimension

A **coordinate system** for a $d$-dimensional space, is a procedure that assigns points in the space to $d$ values so that no distinct points are assigned the same set of $d$ values. 

The **dimension** of a space is the minimum number of coordinates needed to uniquely specify locations in the space.

:::

Coordinate systems assign locations in a space to lists of numbers. Each number in that list is a coordinate. Distinct coordinates convey different information about the location of the point specified by the coordinates. 

Consider a plane. Points on the plane can be described by:

1. Fixing an origin and two perpendicular axes. Then, measure the distance from the point to the origin on each axis. Call these numbers $x_1$ and $x_2$. Then each pair $[x_1,x_2]$ identifies a point in the plane that is $x_1$ units away from the origin along the first axis, and $x_2$ units away from the origin along the second axis. This is a **Cartesian coordinate system.**

    ![Cartesian Coordinates.](Cartesian_Coordinates.png "Cartesian Coordinates.") 

    - For instance, imagine that you are at a rectangular desk. Place a point in the center of the desk. Align one axis to run straight away from you. This is the front-back axis. Then, align the other axis to run perpendicular to the first axis. This is the left-right axis. Then, a point $[1.5,-0.2]$ is a location $1.5$ units back, and $0.2$ units left of center. Notice that the axes must be oriented. We had to choose that $+1.5$ meant motion backward, not motion forward, and that $-0.2$ meant motion left, not right. 

    - Imagine you are given a rectangular piece of graph paper. Place a dot on one of the corners in the graph paper. Call this the origin. Then each other corner in the grid can be represented with a pair of numbers, $[x_1,x_2]$ corresponding to the distance up, and the distance across, from the origin.

    - Cartesian coordinates generalize easily to higher dimensional spaces. For instance, in three-dimensions, we can specify locations by measuring a distance out, across, and up, from an origin. 

1. Fix an origin and one ray leaving the origin. Then, for any point, draw a line segment from the origin to the point. Measure the length of the line segment, and the angle between the line segment and your reference ray. Now each pair, $[r,\theta]$ represents a point on the plane. This is a **polar coordinate system.**

    ![Polar Coordinates.](Polar_Coordinates.png "Polar Coordinates.") 

    - By convention, the first polar coordinate is usually denoted $r$ for "radius." It is the radius of a circle, centered at the origin, passing through the point of interest.

    - By convention, the first polar coordinate is usually denoted $\theta$ for "angle." Sometimes we denote angles with $\phi$, or, less frequently, $\psi$. 
 

In both cases we started by fixing an **origin**. The origin of a coordinate system is the location assigned coordinates $[0,0]$. 

Coordinate systems assign points in a space to lists of numbers. Since distinct coordinates encode distinct information, their order matters, and we usually fix a convention for their order. For instance, in polar coordinates, it is standard practice to list the radius before the angle. 

## Vectors

Finitely long, ordered lists of numbers are *vectors.*

:::{note} Vector Definition

A **vector** in $d$-dimensions is an ordered list of $d$ numbers. 

For instance, $v = [1,4,-2]$ is a vector in 3 dimensions.

A vector is not a set, since the vectors $[1,4,-2]$ and the vectors $[4,-2,1]$ are distinct. 

:::

For example, $v = [v_1,v_2]$ is a two-dimensional vector with entries $v_1$ and $v_2$. If it is unclear whether $v$ is a vector, then we might use the notation $\vec{v}$ to emphasize that $\vec{v}$ is a vector. 

Often, vectors are interpreted as arrows, emanating from the origin, and terminating at the point in a Cartesian coordinate system with coordinates equal to the entries of the vector. For example, the vector $[2,5]$ is the arrows starting at the origin, $[0,0]$ and ending 2 units across and 5 units up from the origin. The vector $[4,-1]$ is the arrow starting at the origin, $[0,0]$, and ending 4 units across and 1 unit *down* from the origin. 

Here are two example vectors:

![Example Vectors.](Vectors.png "Example Vectors.") 

:::{note} Random Vector Definition

A **random vector** in $d$-dimensions is an ordered list of $d$ random variables. 

:::

In other words, a random vector is just a collection of random numbers.

### Vector Operations

We can do algebra with vectors just like we can do algebra with numbers. The algebra of vectors is the subject of *linear algebra*. We will only need the essentials.

#### Vector Addition

:::{note} Vector Vector Addition

If $v$ and $w$ are both vectors with $d$ entries, then:

$$v + w = \left[\begin{array}{c} v_1 \\ v_2 \\ \vdots \\ v_d \end{array} \right] +  \left[\begin{array}{c} w_1 \\ w_2 \\ \vdots \\ w_d \end{array} \right] = \left[\begin{array}{c} v_1 + w_1 \\ v_2 + w_2 \\ \vdots \\ v_d + w_d \end{array} \right] $$

In other words, $v + w$ is a new vector formed by summing each pair of entries in $v$ and $w$.
:::

For example:

$$\left[\begin{array}{c} 2 \\ -1 \end{array} \right] +  \left[\begin{array}{c} 1 \\ 3 \end{array} \right] = \left[\begin{array}{c} 2 + 1 \\ -1 + 3 \end{array} \right] = \left[\begin{array}{c} 3 \\ 2 \end{array} \right]  $$

We can draw this new vector using the **tip-to-tail** rule. Starting from the origin, move to the location $v$. Then, from $v$, move to $v + w$ by translating the arrow $w$ so that it starts from $v$ instead of from the origin.

![Vector Addition.](Vector_Addition.png "Vector Addition.") 

Alternately, draw a parallelogram with a corner at the origin, and sides parallel to $v$ and $w$. The missing corner is $v + w$. 

Like scalar addition, vector addition is **commutative** and **associative** so:

$$v + w = w + v, \quad (v + w) + u = v + (w + u). $$


#### Scalar Vector Multiplication

:::{note} Scalars Definition

A **scalar** is a single number.

:::

It is common practice to denote scalars with greek letters (e.g. $\lambda$) and vectors with roman letters (e.g. $x$, $y$, $v$). Sometimes scalars are represented with roman letters drawn from the start of the alphabet (e.g. $a$, $b$, $c$). 

:::{note} Scalar Vector Multiplication

The product $\lambda v$ is defined $\lambda [v_1,v_2,...,v_d] = [\lambda v_1,\lambda v_2, ..., \lambda v_d]$. 


:::

For example, if $\lambda = 3$ and $v = [1,-2]$ then:

$$\lambda v = 3 \times \left[\begin{array}{c} 1 \\ -2 \end{array} \right]  = \left[\begin{array}{c} 3 \times 1 \\ 3 \times (-2) \end{array} \right] = \left[\begin{array}{c} 3  \\ -6 \end{array} \right]$$

The result is new vector, parallel to the original vector, but three times longer. 

In general, multiplying a vector by a scalar *scales* the vector by the value of the scalar. Multiplying by 2 doubles the length of the vector. Multiplying by $-1$ reflects the vectors through the origin. It produces a vector pointing in the opposite direction. 

Here are examples for the vectors $v$ and $w$ shown in the previous figures:

![Scaling Vectors.](Vector_Scaling.png "Scaling Vectors.") 

:::{note} Parallel Vectors

Two vectors $v$ and $w$ are **parallel** if and only if $v$ and $w$ are proportional to each other. That is, there must exists some $\lambda$ such that $v = \lambda w$.

:::


#### Linear Combination

:::{note} Linear Combination

A **linear combination** of two vectors, $v$, $w$, is a combination of the form $\lambda v + \mu w$ where $\lambda$ and $\mu$ are scalars.

The result is a new vector formed by:

1. Scaling $v$ by $\lambda$.
1. Scaling $w$ by $\mu$.
1. Adding the scaled vectors tip to tail.

:::

Note, the usual order of operations applies. Multiply by the scalars before adding the vectors.

For example:

$$ 3 \times \left[\begin{array}{c} 1 \\ -2 \end{array} \right] + 2 \times \left[\begin{array}{c} 3 \\ 0 \end{array} \right] = \left[\begin{array}{c} 3 \\ -6 \end{array} \right] + \left[\begin{array}{c} 6 \\ 0 \end{array} \right] =  \left[\begin{array}{c} 9 \\ -6 \end{array} \right].  $$

Here are two linear combinations of the vectors $v$ and $w$ illustrated earlier:

### Vector Geometry

Since vectors may be represented with arrows, every vector has a **length** (magnitude) and **direction**.

#### Length/Magnitude

The length of a vector is found using the Pythagorean theorem. We denote the length of the vector $v$ with $\|v\|$. 

In two dimensions:

$$\|v\| = \sqrt{v_1^2 + v_2^2}. $$

For example:

![The Lengths of Two Vectors.](Norms.png "Lengths of the vectors $v$ and $w$.") 

The same formula generalizes to higher dimensions.

:::{note} Length/Magnitude

The **length** (or **magnitude**) of a $d$-dimensional vector, $v$, is given by:

$$\|v\| = \left( \sum_{j=1}^d v_j^2 \right)^{\frac{1}{2}} $$

:::

In other words, the magnitude of a vector is the square root, of the sum, of its entries squared. To find its magnitude, square every entry, add the squares, and square root the result. 

For example, the magnitude of the vector $[3,4]$ is $\sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5$. 


:::{note} Unit Vectors

A vector $v$ is a **unit vector** if $\|v\| = 1$. 

:::

The only vector of length zero is the **zero vector**: $[0,0,0...0]$. A vector is **nonzero** if at least one of its entries is nonzero.

#### Direction

The direction of a vector may be expressed by standardizing its length. To standarize, we *normalize* vectors:

:::{note}

Given a vector $v$, we can **normalize** it, by defining the unit vector:

$$\hat{v} = \frac{v}{\|v\|} $$

You should confirm that $\|\hat{v}\| = 1$. Since $\|v\|$ is a scalar, $\hat{v}$ is proportional to $v$, so $\hat{v}$ is parallel to $v$. 

:::

By normalizing, we produce a unit vector parallel to our original vector. This unit vector is the same for all vectors parallel to the original vector, so represents the direction of the vector. The unit vector always has length 1, so is independent of the magnitude of the vector. 

Here are the unit vectors for the example vectors $v$ and $w$ illustrated earlier:

![Unit Vectors.](Unit_Vectors.png "Unit Vectors.") 

Two vectors, $v$ and $w$, are parallel if and only their unit vectors are identical, or are negations of each other (e.g. $\hat{v} = -\hat{w}$). Two vectors, $v$ and $w$, point in the same direction if and only if they share the same unit vector, $\hat{v} = \hat{w}$. 

#### Angles and Inner Products

We can measure the angle between two vectors using an *inner product*:

:::{note} Inner Products

Given two $d$-dimensional vectors $v$ and $w$, their inner product is:

$$v \cdot w = v_1 \times w_1 + v_2 \times w_2 + ... v_d \times w_d = \sum_{j=1}^d v_j \times w_j.$$

So, the inner product is the *sum of the entrywise products.* To find the inner product, multiply all the entries, then add the products.

:::

For example, $[3,4] \cdot [-1,2] = 3 \times (-1) + 4 \times 2 = -3 + 8 = 5$. 

Inner products are sometimes called "dot" products for the $v \cdot w$ notation.

We care about inner products since they are also related to the lengths of the two vectors, and the angle between them.

:::{note} Inner Products and Geometry

$$v \cdot w = \|v\| \|w\| \cos(\theta). $$

:::

In other words, the inner product between two vectors equals the product of their lengths with the cosine of the angle between the two vectors. 

We can rearrange this formula to find the angle between two vectors using their inner product.

$$\theta = \cos^{-1}\left(\frac{v \cdot w}{\|v\| \|w\|} \right).$$

The two example vectors we've illustrated throughout are separated by about 45 degrees. You can check using their inner products and lengths.

![Angle Between Two Vectors.](Angle.png "Angle Between Two Vectors.") 


:::{note} Perpendicular Vectors

Two nonzero vectors, $v$ and $w$, are **perpendicular** if the angle between the two vectors is 90 degrees. Perpendicular vectors are **orthogonal**. 

:::

We can use the inner product to test whether two vectors are perpendicular. The cosine of $\pm$90 degrees is zero. The cosine of any other angle is nonzero. So, two vectors are perpendicular if and only if $\cos(\theta) = 0$. This requires $v \cdot w = 0$ since $\|v\| \neq 0$ and $\|w\| \neq 0$ when both are nonzero vectors. 


:::{note} Checking Perpendicularity with Inner Products

Two nonzero vectors $v$ and $w$ are perpendicular if and only if $v \cdot w = 0$. 

:::

For example, the vectors $[1,-1]$ and $[4,4]$ are perpendicular since $1 \times 4 + (-1) \times 4 = 4 - 4 = 0$. The vectors $[1,-1]$ and $[4,6]$ are not perpendicular since $1 \times 4 + (-1) \times 6 = 4 - 6 = -2 \neq 0$. 

## Exchanging Polar and Cartesian Coordinates

### Polar to Cartesian

Suppose that $x$ is a point in a plane with radius $r$ and angle $\theta$ counterclockwise from the positive horizontal axis. Then:

:::{note} Cartesian from Polar

$$x_1 = r \cos(\theta), \quad x_2 = r \sin(\theta) $$

:::

### Cartesian to Polar

Suppose that $x$ is a point in a plane with coordinates $x_1$ and $x_2$. Then:

:::{note} Polar from Cartesian

$$r = \|x\| = \sqrt{x_1^2 + x_2^2}, \quad \theta = \tan^{-1}(x_2/x_1) $$

where $\tan^{-1}$ is the inverse or "arc"-tangent.

:::

The equation for $\theta$ is justified since:

$$\frac{x_2}{x_1} = \frac{r \sin(\theta)}{r \cos(\theta)} = \frac{\sin(\theta)}{\cos(\theta)} = \tan(\theta). $$
