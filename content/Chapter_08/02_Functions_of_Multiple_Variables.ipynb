# Functions of Multiple Variables

Just like random variables are characterized by distribution functions, random vectors can be characterized by distribution functions. Like distribution functions for random variables that accept a possible value, $x$, and return a single output chance ($\text{Pr}(X = x)$, $\text{Pr}(X \leq x)$, etc.), distribution functions for random vectors accept a possible value of the random vector $X$, and return a single output chance. When we work with random variables, the possible values, $x$, are scalars. When we work with random vectors, the possible values are vectors, so the distribution functions must be functions of *multiple* variables. They accept vectors, $x$.

:::{note} Functions of Multiple Variables

A function $f$ is a **function of multiple variables** if it accepts **vectors** as its input. It is **scalar valued** if it returns a scalar. It is **vector valued** if it returns a vector.

:::

For example:

1. $f(x) = x_1 + x_2$ is a scalar-valued function of bivariate inputs since it accepts vectors with two entries, and returns scalars.

1. $g(x) = x_1 x_2 - x_3 x_4$ is a scalar-valued function that accepts vectors with four entries. 

1. $h(x) = [x_1 + x_2, x_1 x_2 - x_3 x_4]$ is a vector valued function of multiple variables since it returns a vector with two entries from a vector with four entries. Every vector-valued function can be represented as a collection of scalar valued functions: $h(x) = [h_1(x),h_2(x)]$. In this example, $h_1(x) = f(x)$, $h_2(x) = g(x)$.  

We will start by focusing on the simplest case. Scalar valued function of two variables. A scalar valued function of two variables is a *surface.*

:::{note} Surfaces

Scalar valued functions of multiple variables define **surfaces**. You can imagine a scalar valued function of two variables as a landscape. The input coordinates correspond to poisitions in a plane. The output value corresponds to the height of the landscape at that position. 

:::

When working with functions on two variables, it is convention to denote the pair of input variables $[x,y]$ and the output variable $z$ so $f(x,y)$ returns a height on the $z$ axis. Typically, $x$ is represented with a horizontal axis, $y$ is represented with an axis pointing out of the page (imagine, and axis pointing directly at you, perpendicular to your computer screen), and $z$ is represented with a vertical axis.

## Visualizing Surfaces ...

Here are three distinct ways to visualize surfaces:

### ... Using Cross Sections

Suppose that $f(x)$ is a scalar valued function on two variables. To define a **cross-section**, fix one of the input variables, and vary the other. For example, we can produce cross-sections by defining a new function $g(x)$ where $g(x) = f(x,y_*)$ for some fixed value $y = y_*$. Varying $y_*$ produces different cross-sections. We could call these "$x$-cross section" since they are cross-sections produced by fixing the input $y$, and varying the input $x$. These are scalar valued functions of $x$. 

Visually, you can imagine each $x$ cross section as the curve produced by starting from some $[x,y_*]$, then cutting the surface in the horizontal direction. 

The video below shows two different $x$ cross sections of a surface. Each correspond to a different fixed value of $y$. In this video the x-axis is oriented out of the screen.

To build up the full surface, we can generate a collection of x cross sections and y cross sections. Plotting both creates a mesh, or net, whose height corresponds to the height of the surface. The mesh shown on the full surface plot is a collection of cross sections.

Cross section plots are useful if we want to emphasize how the height of a surface varies as we vary a single coordinate at a time. They are expecially useful in probability since the cross sections of a distribution function are proportional to conditional distributions. 

### ... Using a Heatmap

We can also avoid drawing in three dimensions by coloring the surface according to its height. If we assign each height a different color, then we can view the surface from above, and use the colors to read off the height of the surface at each $[x,y]$ pair. For instance, we could use a color bar that assigns blue to large positive values of $z$, and gold to large negative values of $z$.

### ... Using Level Sets

We can also visualize surfaces by studying the collection of all inputs, $[x,y]$, that return a fixed value. These are all points in the plane where the function $f(x,y)$ returns the same value, or, where the surface attains the same height. We call these collections, *level sets.*

:::{note} Level Sets

A **level set** of a surface $f$ is the collection of all inputs such that the surface returns the same output value. The $c$ level set is the collection: $\{[x,y] \text{ such that } f(x,y) = c\}$

:::

You can imagine level sets as the curve produced by intersecting $f(x,y)$ with the constant function $g(x,y) = c$ for varying $c$. Equivalently, the $c$ level set is the curve produced by cutting the surface in two along a plane parallel to the input plane, with height $c$ on the $z$ axis.

:::{note} Contour Plots

A **contour plot** is a plot produced by drawing the level sets of a surface for varying input heights. Contour plots may be drawn in three dimensions, or viewed "from above" in a top-down view. When viewed from above, a contour plot represents a three dimensional surface using only a two-dimensional plot. Contours may be colored by the height of the surface, or may be labelled by their height.

:::

Notice, if you've ever read a topographic map, which illustrates the elevation of land using level sets, then you've read a contour plot. Contour plots are very useful since they let us draw surfaces without drawing in three dimensions. 

Often, the best way to visualize a surface is with a combination of a heatmap, and a contour plot. The color provides an immediate visual reference for height, while the contours provide precise visualization of the surface's shape.
