# Derivatives of Multivariate Functions

## Chapter Overview

### Where We Are

Chapters 8 introduced the coordinates used to represent, and the distributions used to sample, random vectors. In particular, [Section 8.2](#ch8.2) and [8.3](#ch8.3) introduced scalar-valued functions of multiple variables. 

### Where We're Going

In this chapter, we will study how to differentiate functions of multiple variables. We will use these tools to find maxima and minima of surfaces, and to find modes of multivariate distributions. 

We will:

1. Define a **partial derivative** as a derivative of a cross-section of a surface and will show how to use partial derivatives to build **linear approximations** to surfaces using their **tangent planes** (see [Section 9.1](#ch9.1)).

1. Show how to summarize the linearization with **gradients** (see [Section 9.2](#ch9.2)). We will:

    - Practice computing gradients

    - Show how to recover **directional derivatives** from gradients

    - Show that the gradient vector points in the direction of steepest ascent, and is perpendicular to the level sets.

1. Combine these ideas to **optimize** surfaces. We will use gradients to:

    - Optimize surfaces without any constraints on the inputs by setting the gradient to zero (see [Section 9.3](#ch9.3)).

    - Optimize surfaces iteratively by stepping in the direction of the gradient, that is, by **gradient ascent** (see [Section 9.3](#ch9.3)).

    - Optimize surfaces subject to constraints on the inputs using **Lagrange multipliers** (see [Section 9.4](#ch9.4)).
